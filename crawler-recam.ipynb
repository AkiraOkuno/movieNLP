```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
letters = '123ABCDEFGHIJKLMNOPQRSTUVWX'

url_init = "https://www.recam.org/?do=db&f="
url_end="&order_by=db_movies.title_pt&country=&title=&time="
```

```python
soup.find_all('div',{'class':'billboardList'})[1].find('h5').text
```

```python
lista = []

for char in letters:
    
    print(char)
    print('------------------------------------------------------------------')
    url = url_init+char+url_end
    resp = requests.get(url)
    soup = BeautifulSoup(resp.content, "html.parser")
    
    for movie in soup.find_all('div',{'class':'billboardList'}):
        
        name = movie.find('h5').text
        print(name)
        ano = movie.find_all('h6')[1].text
        pais = movie.find_all('h6')[2].text

        link = 'https://www.recam.org/' + movie.find('div').find('a')['href']
        resp2 = requests.get(link)
        soup2 = BeautifulSoup(resp2.content, "html.parser")

        descricao = soup2.find('p').text
        duracao = soup2.find_all('tr')[4].find('td').text
        genero = soup2.find_all('tr')[6].find('td').text
        diretor = soup2.find_all('tr')[7].find('td').text
        protagonistas = soup2.find_all('tr')[8].find('td').text

        lista.append([name,ano,pais,descricao,duracao,genero,diretor,protagonistas])
```

```python
len(lista)
```

```python
df= pd.DataFrame(lista)
df.to_csv('filmes_full.csv')
```

```python
dfm=df[[0,3]]
dfm.columns=['nome','descricao']
```
